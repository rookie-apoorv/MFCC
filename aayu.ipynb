```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
xy = 8000  # Constant in the demand function (x * y = 8000)
x_values = np.linspace(1, 20, 100)  # Quantity range from 1 to 20
y_values = xy / x_values  # Calculate corresponding prices

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="HongKong Noodles", color="blue")

# Labels and title
plt.title("Akshika's Demand Function")
plt.xlabel("Quantity demanded per semester")
plt.ylabel("Price")
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
